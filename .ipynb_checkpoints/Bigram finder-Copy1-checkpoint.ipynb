# Find Bigram in text datasets

```python
#import required packages
import nltk as nk
import pandas as pd
import re
from collections import Counter
import csv
```

```python
#read dataset
data = pd.read_csv('tweets.csv')
data.head(10)
```

```python
#find top retweeted tweets about alhabib hospital
top_retweeted = data.sort_values(by=['retweets_count'],ascending=False).head(5)[['tweet','username']]
#store the result externally
top_retweeted.to_csv('top_retweeted.csv',sep=',', encoding='utf-8-sig')
top_retweeted
```

Bigram Extraction

```python
#define a function that tokenizes sentences to bigrams
def bigrams(sentence):
    text = re.sub('\W', ' ', sentence.lower())
    words = text.split()
    return zip(words, words[1:])
```

```python
#Day1
#initiate a counter to count the distribution of the bigrams
counts_1 = Counter()
for line in data['tweet']:
    counts_1.update(bigrams(line))
#convert dictionary to series
counts_series_1= pd.Series(counts_1, name='Bigram_Count')
counts_series_1.index.name = 'Count'
#create a dataframe
bigram_count=pd.DataFrame(counts_series_1.reset_index())
#create a new column that specifies the date 
bigram_count.head()
```

```python
# changing columns names using .columns() 
bigram_count.columns = ['Word1', 'Word2', 'Bigram_Count'] 
#merge two words in one column
bigram_count['bigram'] = bigram_count['Word1']+" "+bigram_count['Word2']
bigram_count_ordered = bigram_count.sort_values(by=['Bigram_Count'],ascending=False)
bigram_count_ordered.head()
```

```python
#export dataframe as a csv file
bigram_count_ordered.to_csv('habib.csv',sep=',', encoding='utf-8-sig')
```
